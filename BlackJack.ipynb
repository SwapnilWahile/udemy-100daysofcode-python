```python
import random
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
usrc = []
def user():
    for i in range(1, 3):
        user_car = cards[random.randint(0,12)]
        usrc.append(user_car)
    print(f"your cards {usrc}")
    addu = usrc[0] + usrc [1]
    return addu

def drawu_card():
        dous_car = cards[random.randint(0,12)]
        usrc.append(dous_car)
        print(f"your drawn card {dous_car}")
        return dous_car
        
dlc = []
def dealer():
    for i in range(1, 3):
        dl_car = cards[random.randint(0,12)]
        dlc.append(dl_car)
    print(f"Dealer card {dlc[0]}")
    addd = dlc[0] + dlc[1] 
    return addd

def draw_card():
        dodl_car = cards[random.randint(0,12)]
        dlc.append(dodl_car)
        print(f"Dealer drawn card {dodl_car}")
        #adddl = dlc[0] + dlc[1] + do
        return dodl_car
   

dec = input("Do you wants to play blackjack: y or n ")

if dec == "y":
    user()
    dealer()    
    continu = True 
    dlcam = 0
    usercam = 0
    def winner_dec():
        if usercam == 21 and dlcam == 21:
            print(f"Dealers all cards {dlc}")
            print("Dealer is winner ")
        elif usercam == 21 or dlcam == 21:
            print(f"Dealers all cards {dlc}")
            print("Its blackjack game ends ")
        elif usercam > 21:
            print(f"Dealers all cards {dlc}")
            print("You lost game ends")
        elif dlcam > 21:
            print(f"Dealers all cards {dlc}")
            print("You won")
        elif usercam > dlcam:
            print(f"Dealers all cards {dlc}")
            print("you are winner ")
        elif dlcam > usercam:
            print(f"Dealers all cards {dlc}")
            print("Dealer is winner ")
        else:
            print(f"Dealers all cards {dlc}")
            print("Draw")
    while continu:    
        storhit = input("do you want to stand or hit ")
        if storhit == "hit":
            drawu_card()
            for x in usrc:
                usercam += x
            if usercam == 20 or usercam > 20:
                    winner_dec()
                    continu = False
        else:
            if dlcam <= 16:
                should_draw = True
                while should_draw:
                    draw_card()
                    for z in dlc:
                        dlcam += z
                    if dlcam > 16:
                        should_draw = False
                        winner_dec()
                        continu = False
            else:
                winner_dec()
else:
    print("bye")
```
